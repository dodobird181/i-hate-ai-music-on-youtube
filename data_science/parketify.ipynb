# Parketify Script
Turns a CSV file with a string-embedding column into a .parket file where each embedding dimension is split into its own column.

```python
CSV_FILENAME = "tmp_data/sam_test_video_title_data.csv"
EMBEDDING_COLUMN_NAME = "title_embedding"
```

```python
import pandas as pd
df = pd.read_csv(CSV_FILENAME)
```

```python
data = []
for i, row in df.iterrows():
    embeddings = row[EMBEDDING_COLUMN_NAME].replace("\n", "").replace("[", "").replace("]", "").split()
    row_data = {f"{EMBEDDING_COLUMN_NAME}_{j}": float(embedding) for j, embedding in enumerate(embeddings)} | {"label": row["label"]}
    data.append(row_data)
df_split = pd.DataFrame(data)
df_split
```

```python
parket_filename = f'{CSV_FILENAME.split(".")[0]}.parquet'
df_split.to_parquet(parket_filename)
```
